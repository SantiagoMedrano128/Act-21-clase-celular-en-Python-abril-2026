```python
class Celular:
    # Método constructor para inicializar atributos
    def __init__(self, marca, modelo, precio):
        self.marca = marca
        self.modelo = modelo
        self.precio = precio
        self.encendido = False

    # Método para mostrar información
    def mostrar_info(self):
        return f"{self.marca} {self.modelo} - ${self.precio}"

    # Método para encender el celular
    def encender(self):
        self.encendido = True
        print(f"El {self.modelo} está encendido.")

# Instanciar el objeto
mi_celular = Celular("Samsung", "Galaxy S21", 799.99)
su_celular = Celular("Xiaomi", "Redmi Note 10", 399.99)

# Acceder a los atributos
print("celular 1")
print(mi_celular.mostrar_info())
mi_celular.encender()
print("\n" + "-"*30 + "\n")
# Lamada  a las funciones para cada objeto celular 1 y celular 2
print("Celular 2")
print(su_celular.mostrar_info())
su_celular.encender()
print("\n" + "-"*30 + "\n")
print("Programa terminado")
print("Creador de software por: Santiago Medrano NL = 37")
```
